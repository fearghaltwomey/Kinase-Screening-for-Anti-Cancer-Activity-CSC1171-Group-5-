```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame({"Fingerprint": [[1, 0, 1, 0], [0, 0, 0, 1], [0, 1, 1, 0], [1, 1, 1, 1], [0, 0, 0, 0]], "Class": [1, 1, 0, 1, 0]})
df["Value"] = 0
```

```python
# X/y split
# Assuming df is concat of all data
X = df["Fingerprint"]
y = df["Class"]
# OR
# y = df["Value"]
```

```python
X_remainder, X_test, y_remainder, y_test = train_test_split(X, y, test_size= 0.2)
X_train, X_val, y_train, y_val = train_test_split(X_remainder, y_remainder, test_size= 0.25)

# can call:
# X_train, y_train
# X_test, y_test
# X_val, y_val
```

```python
# Functionalised
# mode: 1 = clf, 0 = reg
def splits(df, mode= 1):
    if df.shape[0] <= 1000:
        raise ValueError(f"Not enough data: {df.shape[0]} datapieces")
    elif list(df) != ["SMILES", "Class", "Value", "Kinase", "Group", "Fingerprint"]:
        raise TypeError(f"Incorrect dataframe structure: {list(df)}")  
    else:
        if mode == 1:
            y = df["Class"]
        elif mode == 0:
            y = df["Value"]
        else:
            raise ValueError(f"{mode} is not a valid key. Use mode= 1 (clf) or mode= 0 (reg)")
        X = df["Fingerprint"]
        X_remainder, X_test, y_remainder, y_test = train_test_split(X, y, test_size= 0.2)
        X_train, X_val, y_train, y_val = train_test_split(X_remainder, y_remainder, test_size= 0.25)

        return X_train, X_test, X_val, y_train, y_test, y_val
```

```python
list(df)
splits(df)
```
