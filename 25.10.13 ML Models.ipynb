```python
# Import functions
from lightgbm import LGBMClassifier, LGBMRegressor
from xgboost import XGBClassifier, XGBRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.svm import SVC, SVR
```

```python
# Random Lists for Proof-of-Concept
X_train = [[1, 0, 0, 1], [0, 1, 1, 1]]
y_train = [1, 0]
```

```python
# CLF / REG assignment

# Light Gradient Boosting (LGB)
clfLGB = LGBMClassifier()
regLGB = LGBMRegressor()

clfLGB.fit(X_train, y_train)
regLGB.fit(X_train, y_train)

# Extreme Gradient Boosting (XGB)
clfXGB = XGBClassifier()
regXGB = XGBRegressor()

clfXGB.fit(X_train, y_train)
regXGB.fit(X_train, y_train)

# Multilayer Perceptron (MLP)
clfMLP = MLPClassifier()
regMLP = MLPRegressor()

clfMLP.fit(X_train, y_train)
regMLP.fit(X_train, y_train)

# Random Forest (RF)
clfRF = RandomForestClassifier()
regRF = RandomForestRegressor()

clfRF.fit(X_train, y_train)
regRF.fit(X_train, y_train)

# Support Vector Machine (SVM)
clfSVM = SVC()
regSVM = SVR()

clfSVM.fit(X_train, y_train)
regSVM.fit(X_train, y_train)
```
